```python
# agnesi_witch.ipynb

# Cell 1

from __future__ import annotations

import typing
import matplotlib.pyplot as plt
import numpy as np

if typing.TYPE_CHECKING:
    from numpy.typing import NDArray
    from matplotlib.axes import Axes
    
%matplotlib widget    
"""This is mostly taken from Dr. Biersach's solution"""
def f(x: NDArray[np.float_]) -> NDArray[np.float_]:
    return 1 / (1 + x ** 2)


def p_f(x: float, num_terms: int) -> float:
    y: float = 0.0
    for n in range(num_terms):
        y += (-(x ** 2)) ** n
    return y


x: NDArray[np.float_] = np.linspace(-1.5, 1.5, 1000)
fig: Figure = plt.figure(" ")
ax: Axes = fig.add_subplot()
ax.plot(x, f(x), label = "Exact")

v = np.vectorize(p_f, excluded = ("num_terms"))
for n in range(2, 8):
    ax.plot(x, v(x, num_terms = n), label = f"{n} terms")

    ax.set_title("Runge's Phenomenon (Witch of Agnesi)")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.legend(loc = "best")
    
    ax.grid()
    ax.set_ylim(-2, 2)
    
    ax.axvline(x = -1, color = "black", linestyle = "--", linewidth = 2)
    ax.axvline(x = 1, color = "black", linestyle = "--", linewidth = 2)

plt.show()
```

```python
# Cell 2

"""This function diverges as the power series approach -1 and 1 becuase the funciton is undefined at -1 and 1. Increasing the power won't change that in any way.

   This demonstrates Runge's Phenomenon by showing that the power series explodes as the powers increas and they get increasingly further from the original function."""
```
